```python
# read the events file

events_df = pd.read_json("/Users/oreogundipe/Downloads/events.json")
events_df
```

```python
import pandas as pd
for event in events_df.iterrows():

    print(event[1]['startTimestamp'])
    timestamp = event[1]['startTimestamp']

    try:
        eeg_dataset_df = pd.read_json(f"/Users/oreogundipe/lab/neuroscripts/logger/data/2022/ore/rawBrainwaves_{timestamp}.json")
        eeg_dataset_df['unixTimestamp'] = eeg_dataset_df[['unixTimestamp', 'index']].apply(lambda x: x['unixTimestamp'] + (4*x['index']), axis=1)
        eeg_dataset_df.drop('index', axis=1, inplace=True)

        # ore's format
        eeg_dataset_df.to_csv(f"/Users/oreogundipe/lab/neuroscripts/logger/data/processed/rawBrainwaves_{timestamp}.csv")

        # format to alex's tables
        eeg_dataset_df = eeg_dataset_df.rename(columns={
            "CP3": "Channel 1",
            "C3": "Channel 2",
            "F5": "Channel 3",
            "PO3": "Channel 4",
            "PO4": "Channel 5",
            "F6": "Channel 6",
            "C4": "Channel 7",
            "CP4": "Channel 8",
            "unixTimestamp": "lsl_time_stamp"
        })

        eeg_dataset_df["Recording Timestamp Base (Unix Epoch)"] = ""
        eeg_dataset_df["Recording Timestamp Remainder"] = ""

        eeg_dataset_df.to_csv(f'/Users/oreogundipe/lab/neuroscripts/logger/data/processed/Crown-853_{timestamp}.csv')
    except Exception:
        print(f"error processing for {timestamp}")

```

```python
# display line plot for days recordings existed
# split into morning, afternoon, evening
import pandas as pd
from pytz import timezone


def add_timeofday_column(df):
    pst = timezone('America/Vancouver')
    df['convertedTimestamp'] = pd.to_datetime(df['startTimestamp'], unit='s', utc=True).dt.tz_convert(pst)
    
    # Map hours to time of day categories
    df['timeofday'] = df['convertedTimestamp'].dt.hour.apply(lambda x: "morning" if 6 <= x < 12 else ("afternoon" if 12 <= x < 18 else "evening"))
    return df

timed_df = add_timeofday_column(events_df)
```
